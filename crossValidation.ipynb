```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
```

```python
heart = pd.read_csv("/content/heart.csv")
```

```python
X= heart.drop(columns='target', axis=1)
Y= heart['target']
```

```python
X
```

```python
Y
```

```python
X_train,X_test, Y_train,Y_test= train_test_split(X, Y, stratify=Y, test_size=0.2, random_state=3)
```

```python
X_train
```

```python
print(X_train.shape,Y_train.shape, X_test.shape, Y_test.shape)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
# comparing the models
# list of models
models= [LogisticRegression(max_iter=1000), SVC(kernel='linear'), KNeighborsClassifier(), RandomForestClassifier()]
```

```python
def compare_models_train_test_acc():

  for model in models:
    model.fit(X_train,Y_train)

    Y_pred= model.predict(X_test)

    accuracy= accuracy_score(Y_test,Y_pred)

    print("Accuracy of the ", model, "is ", accuracy )
```

```python
X_test.shape
```

```python
compare_models_train_test_acc()
```

```python
# cross validation-- LogisticRegression

cv_score_lr= cross_val_score(LogisticRegression(max_iter=1000),X,Y,cv=5)

print(cv_score_lr)
mean_lr = sum(cv_score_lr)/len(cv_score_lr)
mean_lr= mean_lr*100
mean_lr= round(mean_lr,2)
print(mean_lr)
```

```python
# cross validation-- SVM

cv_score_svm= cross_val_score(SVC(kernel='linear'),X,Y,cv=5)

print(cv_score_svm)
mean_svm = sum(cv_score_svm)/len(cv_score_svm)
mean_svm= mean_svm*100
mean_svm= round(mean_svm,2)
print(mean_svm)
```

```python
# cross validation function for model comparision
# list of models
models= [LogisticRegression(max_iter=1000), SVC(kernel='linear'), KNeighborsClassifier(), RandomForestClassifier()]
```

```python
def compare_models_cross_val_score_acc():

  for model in models:
    cv_score= cross_val_score(model,X,Y,cv=5)
    mean_model = sum(cv_score)/len(cv_score)
    mean_model= mean_model*100
    mean_model= round(mean_model,2)
    print("Cross validation score for the ",model, "is " ,cv_score)
    print("Accuaracy of the ",model, "is", mean_model)
    print("-----------------------------------------------------")
```

```python
compare_models_cross_val_score_acc()
```
